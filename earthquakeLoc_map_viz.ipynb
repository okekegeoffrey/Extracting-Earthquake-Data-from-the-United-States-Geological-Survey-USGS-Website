```python
# import necessary packeges
import folium
import pandas as pd
```

```python
# load the csv data
df = pd.read_csv("earthquakes_data.csv")
df.head()
```

```python
# check the no. of rows
df.shape
```

```python
# create empty map/ base map and center it around the mean values of lat and long
map_center = [df['Latitude'].mean(),df['Longitude'].mean()]
m = folium.Map(location=map_center,zoom_start=2)
```

```python
# add earthquake datapoints to the map
for index, row in df.iterrows():
    folium.CircleMarker(
        location=[row['Latitude'],row['Longitude']],
        radius=row['Magnitude'],
        popup=row['Location'],
        color='red',
        fill=True,
        fill_opacity=0.7
    ).add_to(m)
```

```python
# save map in html file or display on jupyter
m.save('earthquake_map.html')
m
```
